```python
from pyensembl import cached_release
```

```python
release_to_id_to_name = {}
for release in [75, 76, 77, 78, 79, 80, 81, 82, 83, 84,  87, 92, 93]:
    genome = cached_release(release)
    print(genome)
    valid_gene_ids = set(genome.gene_ids())
    id_to_name = {x: genome.gene_by_id(x).name for x in tcga if x in valid_gene_ids}
    release_to_id_to_name[release] = id_to_name
```

```python
{release: len(d) for (release, d) in release_to_id_to_name.items()}
```

```python
from collections import defaultdict
combined_id_to_name = defaultdict(set)
for r in reversed(sorted(release_to_id_to_name.keys())):
    id_to_name = release_to_id_to_name[r]
    for x in tcga:
        if x in id_to_name:
            combined_id_to_name[x].add(id_to_name[x])
all_names = set([])
for names in combined_id_to_name.values():
    for name in names:
        all_names.add(name)
```
